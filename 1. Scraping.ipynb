L'objectif de cette partie est d'obtenir une première base de données à partir de scraping d'un site choisi. <br/> Pour notre analyse, nous avons choisi un site d'où nous récupérons des données relatives au logement disponibles à la location. <br/> <br/>
**Nexity** est un site qui propose des offres d'achat, de location ou de vente de biens immobiliers (appartements, maisons, terrains, immeubles, ...) principalement en `France`.  <br/> 
Les caractéristiques ou encore les filtres appliqués à la recherche sur le site sont les suivants : 
- Appartement, maison
- Disponible pour location

Veuillez noter que puisqu'il s'agit d'un site de logement, les informations varient selon la disponibilité des logements. Aussi,  nous travaillons avec des données collectées le _**24/01/2023**_. Il est donc possible que si ce script est exécuté à nouveau, la base de données obtenue soit différente de celle que nous utilisons pour la suite de notre analyse. Pour éviter tout problème dans la suite de notre analyse nous modifierons le nom de base exportée pour ne pas l'écraser. <br/>

# Initialisation

## Import des modules

```python
from lxml import html
import requests
import urllib
from bs4 import BeautifulSoup
import pandas as pd
```

## Initialisation des listes

```python
Types         = []  # Type de logement (meublé ou non, résidence étudiante ou non)
Titre         = []  # Titre de la page
Logement      = []  # Type de logment(Studio, appartement, ...)
Code_postal   = []  # Code postal du logement
Departement   = []  # Département du logement
Superficie    = []  # Superficie du logement
Essentiel     = []  # Essentiel
Amenagement   = []  # Amenagement
Prix_loyer    = []  # Prix du loyer
Description   = []  # Description du loyer
Loyer_Charge  = []  # details_titre
Copropriete   = []  # details_titre
Bareme        = []  # bareme
Details       = []  # Détails
```

# Boucle sur toutes les pages

```python
for i in range (33):
    if i==1:
        url  = 'https://www.nexity.fr/annonces-immobilieres/location/immobilier/tout/france?types_bien%5B0%5D=appartement&types_bien%5B1%5D=maison'
    else:
        url  = 'https://www.nexity.fr/annonces-immobilieres/location/immobilier/tout/france?page='+str(i)+'&types_bien%5B0%5D=appartement&types_bien%5B1%5D=maison'
        
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    print(i, page.status_code)
    
    blocks = soup.find_all("h2", class_="informations")
    
    links=[]
    for block in blocks:
        link=block.find("a", target="_blank")['href']
        links.append('https://www.nexity.fr' + link)
        

    for link in links:   
        page = requests.get(link)
        soup = BeautifulSoup(page.content, "html.parser")
        print(page.status_code)

        # Types
        Types_ = soup.find_all("div", class_="flap flap--not-new")
        if len(Types_) ==0:
            Types.append("NA")
        else:
            Types.append(Types_[0].get_text(strip=True))

        # Titre
        Titre_ = soup.find_all("div", class_="product--title")
        if len(Titre_) ==0:
            Titre.append("NA")
        else:
            Titre.append(Titre_[0].get_text(strip=True)) 

        # logement
        Logement_ = soup.find_all("h2", class_="text--small--type--piece")
        if len(Logement_) ==0:
            Logement.append("NA")
        else:
            Logement.append(Logement_[0].get_text(strip=True))

        # Code_postal
        Code_postal_ = soup.find_all("span", class_="zipcode")
        if len(Code_postal_) ==0:
            Code_postal.append("NA")
        else:
            Code_postal.append(Code_postal_[0].get_text(strip=True))

        # Departement
        Departement_ = soup.find_all("span", class_="city")
        if len(Departement_) ==0:
            Departement.append("NA")
        else:
            Departement.append(Departement_[0].get_text(strip=True))

        # Superficie
        Superficie_ = soup.find_all("h2", class_="text--small--surface--not-new")
        if len(Superficie_) ==0:
            Superficie.append("NA")
        else:
            Superficie.append(Superficie_[0].get_text(strip=True))

        # Essentiel
        Essentiel_ = soup.find_all("div", class_="details")
        if len(Essentiel_) ==0:
            Essentiel.append("NA")
        else:
            Essentiel.append(Essentiel_[0].get_text(strip=True))

        # Amenagement
        Amenagement_ = soup.find_all("div", class_="details")
        if len(Amenagement_) ==0:
            Amenagement.append("NA")
        else:
            Amenagement.append(Amenagement_[1].get_text(strip=True))

        # Loyer_Charge
        Loyer_Charge_ = soup.find_all("div", class_="details")
        if len(Loyer_Charge_) ==0:
            Loyer_Charge.append("NA")
        else:
            Loyer_Charge.append(Loyer_Charge_[2].get_text(strip=True))

        # Copropriete
        Copropriete_ = soup.find_all("div", class_="details")
        if len(Copropriete_) ==0:
            Copropriete.append("NA")
        else:
            Copropriete.append(Copropriete_[3].get_text(strip=True))

        # Prix_loyer
        Prix_loyer_ = soup.find_all("h2", class_="text text--small text--hightlighted text--hightlighted--not-new")
        if len(Prix_loyer_) ==0:
            Prix_loyer.append("NA")
        else:
            Prix_loyer.append(Prix_loyer_[0].get_text(strip=True))

        # Description
        Description_ = soup.find_all("div", class_="description")
        if len(Description_) ==0:
            Description.append("NA")
        else:
            Description.append(Description_[0].get_text(strip=True))

        # Bareme
        Bareme_ = soup.find_all("div", class_="block-characteristiques--bareme")
        if len(Bareme_) ==0:
            Bareme.append("NA")
        else:
            Bareme.append(Bareme_[0].get_text(strip=True))

print(len(Types))        
print(len(Titre))
print(len(Logement))
print(len(Code_postal))
print(len(Departement))
print(len(Superficie))
print(len(Essentiel))
print(len(Amenagement))
print(len(Prix_loyer))
print(len(Description))
print(len(Loyer_Charge))
print(len(Copropriete))
print(len(Bareme))
```

## Enregistrement dans un DataFrame

```python
df = pd.DataFrame({'Types':Types,
                   'Titre':Titre,
                   'Logement':Logement,
                   'Code_postal':Code_postal,
                   'Departement':Departement,
                   'Superficie':Superficie,
                   'Essentiel':Essentiel,
                   'Amenagement':Amenagement,
                   'Prix_loyer':Prix_loyer,
                   'Description':Description,
                   'Loyer_Charge':Loyer_Charge,
                   'Copropriete':Copropriete,
                   'Bareme':Bareme}
                   )
```
